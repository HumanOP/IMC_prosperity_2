```python
weather_data=pd.read_csv('delhi 2024-02-21 to 2024-04-01.csv')
```

```python
weather_data
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import matplotlib.pyplot as plt


# Preprocess data
# Assume 'rain_today' is the target variable
features = weather_data[['temp', 'humidity', 'windspeed']]
target = weather_data['cloudcover']

# Normalize features
features = (features - features.mean()) / features.std()

# Split data into training and testing
train_size = int(len(features) * 0.8)
train_features, test_features = features[:train_size], features[train_size:]
train_target, test_target = target[:train_size], target[train_size:]

# Reshape features for LSTM input
train_features = np.array(train_features).reshape(train_features.shape[0], 1, train_features.shape[1])
test_features = np.array(test_features).reshape(test_features.shape[0], 1, test_features.shape[1])

# Build LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(train_features.shape[1], train_features.shape[2])),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(train_features, train_target, epochs=50, validation_data=(test_features, test_target), verbose=1)

# Plot training history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
